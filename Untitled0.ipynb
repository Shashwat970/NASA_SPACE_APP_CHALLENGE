## Load and analyze data

### Subtask:
Load the `pm25-air-pollution.csv` dataset and perform initial analysis to understand its structure, content, and potential issues.


**Reasoning**:
Load the dataset into a pandas DataFrame and perform initial analysis steps as requested in the instructions.



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/pm25-air-pollution.csv')

# Display the first 5 rows
display(df.head())

# Print column names and data types
print(df.info())

# Display descriptive statistics for numerical columns
display(df.describe())
```

## Prepare data for modeling

### Subtask:
Clean and preprocess the data for time series analysis and prediction. This may include handling missing values, converting data types, and feature engineering.


**Reasoning**:
Handle missing values in the 'Code' column and rename the PM2.5 concentration column as instructed.



```python
# Handle missing values in the 'Code' column by dropping rows with missing codes
df.dropna(subset=['Code'], inplace=True)

# Rename the PM2.5 concentration column
df.rename(columns={'Concentrations of fine particulate matter (PM2.5) - Residence area type: Total': 'PM2.5 Concentration'}, inplace=True)

# Display the first few rows to verify changes
display(df.head())

# Check for missing values again to confirm removal
print(df.isnull().sum())

# Check data types to ensure 'Year' is still integer
print(df.info())
```

## Build and train ml model

### Subtask:
Choose an appropriate time series model (e.g., ARIMA, Prophet, or a deep learning model) and train it on the prepared data to predict future air pollution levels for each country.


**Reasoning**:
Iterate through each country, prepare the data as a time series, initialize and train a Prophet model, and store the trained models in a dictionary.



```python
from prophet import Prophet
import warnings

# Suppress specific warnings from Prophet
warnings.filterwarnings('ignore', message='.*using dates before the start of the time series.*')
warnings.filterwarnings('ignore', message='.*Optimization terminated abnormally.*')
warnings.filterwarnings('ignore', message='.*Fitting a model with separate holiday effects requires.*')


# Dictionary to store trained models for each country
country_models = {}

# Iterate through each unique country
for country in df['Entity'].unique():
    print(f"Training model for {country}...")
    # Filter data for the current country
    country_df = df[df['Entity'] == country].copy()

    # Prepare data for Prophet: rename columns to 'ds' (datestamp) and 'y' (value)
    country_df['Year'] = pd.to_datetime(country_df['Year'], format='%Y')
    country_df = country_df.rename(columns={'Year': 'ds', 'PM2.5 Concentration': 'y'})

    # Ensure there are enough data points for Prophet (Prophet recommends at least 2 years of daily data,
    # but for yearly data, we'll use a heuristic of at least 5 data points)
    if len(country_df) < 5:
        print(f"Skipping {country} due to insufficient data ({len(country_df)} data points).")
        continue

    # Initialize Prophet model
    # Using yearly seasonality as the data is yearly
    model = Prophet(yearly_seasonality=True)

    try:
        # Train the model
        model.fit(country_df)

        # Store the trained model
        country_models[country] = model
        print(f"Model trained successfully for {country}.")
    except Exception as e:
        print(f"Error training model for {country}: {e}")
        continue

print("Finished training models for all countries.")
```

## Generate predictions

### Subtask:
Use the trained model to predict future air pollution levels for each country.


**Reasoning**:
Create future dataframes and generate predictions for each country using the trained Prophet models.



```python
# Dictionary to store future dataframes for each country
country_future = {}

# Iterate through the trained models
for country, model in country_models.items():
    # Create a future dataframe for 10 years
    future = model.make_future_dataframe(periods=10, freq='YS')
    country_future[country] = future

# Dictionary to store predictions for each country
country_predictions = {}

# Iterate through the trained models and their corresponding future dataframes
for country, model in country_models.items():
    # Generate predictions using the future dataframe
    forecast = model.predict(country_future[country])
    country_predictions[country] = forecast

print("Finished generating future dataframes and predictions for all countries.")
```

## Visualize data and predictions

### Subtask:
Create visualizations to show historical data, model predictions, and key trends.


**Reasoning**:
Iterate through each country, filter the historical data and forecast data for the current country, and then plot the historical data and the forecast including the uncertainty interval.



```python
import matplotlib.pyplot as plt
import os

# Create a directory to save the plots if it doesn't exist
if not os.path.exists('plots'):
    os.makedirs('plots')

# Iterate through each country in country_predictions
for country, forecast_df in country_predictions.items():
    # Get the historical data for the current country
    historical_df = df[df['Entity'] == country].copy()
    historical_df['Year'] = pd.to_datetime(historical_df['Year'], format='%Y')

    # Create a plot
    plt.figure(figsize=(12, 6))

    # Plot historical data
    plt.plot(historical_df['Year'], historical_df['PM2.5 Concentration'], label='Historical', marker='o')

    # Plot predicted data and uncertainty interval
    plt.plot(forecast_df['ds'], forecast_df['yhat'], label='Predicted', linestyle='--')
    plt.fill_between(forecast_df['ds'], forecast_df['yhat_lower'], forecast_df['yhat_upper'], alpha=0.3, label='Uncertainty Interval')

    # Add title and labels
    plt.title(f'PM2.5 Concentration Forecast for {country}')
    plt.xlabel('Year')
    plt.ylabel('PM2.5 Concentration')
    plt.legend()
    plt.grid(True)

    # Save the plot as an image
    plt.savefig(f'plots/{country.replace(" ", "_")}_forecast.png')

    # Close the plot to free up memory
    plt.close()

print("Finished generating and saving plots for all countries.")
```
